To show you how to use the features extracted by image awesome, I'm going to run through a quick tutorial of how to build a cat-or-dog classifier.  This approach to building a classifier can be applied to any categories, but is likely to work best for natural image datasets which overlap with the images in the [imagenet database](http://www.image-net.org/).

I'm going to show you how to do this in python, but the same process can be done in any language.

## 1. Data preparation

### 1.1 Upload the images
Download [dogs vs cats dataset](https://www.kaggle.com/c/dogs-vs-cats) from kaggle

extract the images: ```unzip train.zip```

Next you'll need to upload the images to somewhere they can be accessed by image awesome.  Here's a simple python [script for uploading the images to an s3 bucket](https://gist.github.com/shuggiefisher/cd8aa8028738a175f7a2)

```
curl -o s3upload_folder.py https://gist.github.com/shuggiefisher/cd8aa8028738a175f7a2
python s3upload_folder.py my_aws_id my_aws_secret_key mydomain-dogsvscats train
```
Where "mydomain-dogsvscats" is the name of the S3 bucket to create, and "train" is the local folder containing the training images.

### 1.2 Generate a CSV file containing the image URLs

first let's get all the filenames

````python
import os

(sourceDir, dirname, filenames) = os.walk('train').next()
````

now use pandas to make a bzipped CSV file of the image urls

````python
import pandas as pd

df = pd.DataFrame({'filename': filenames})

# prefix the filenames with the s3 bucket url
df['url'] = 'http://s3.amazonaws.com/mydomain-dogsvscats/' + df['filename']

df.head()
````

save it as a compressed CSV

````python
from bz2 import BZ2File

with BZ2File('my_csvfile.csv.bz2', mode='w') as filehandle:
    df.to_csv(
        filehandle,
        columns=['url'],
        header=False,
        index=False
    )
````

## 2.  Upload the CSV file to image awesome for processing

Now we can upload this CSV for processing.  To create a job you'll need a image awesome account, which you can create by signing in with github.  Visit http://www.imageawesome.com/upload and use the form to upload your file.

Image awesome will scan though your CSV, download and resize each image, and run it through a deep convolutional network.  The activity of penultimate layer will be returned as the "features" of the image.

When your job has completed you should receive an email with a link to the results.

## 3. Train classifier on the image awesome results

### 3.1 Create a training and test dataset
The results are stored in HDF5 format. Let's load them, and use them to train our cat-or-dog classifier

````python
fields = ['features', 'processed_urls']

import h5py
with h5py.File('my_results.hdf5') as f:
    data = {field: f[field].value for field in fields}

print data['processed_urls'][:4]
print data['features'][:4]
````

So for each image we now have 4096 float32 values which contain the features extracted by the neural network.  We'll use these features as the input to our cat-or-dog classifier.  First we need to label the images that were processed successfully as a cat or a dog using the filename.

````python
import numpy as np

split_urls = np.char.split(data['processed_urls'], '/')
is_cat = np.char.startswith(np.vstack(split_urls)[:,-1], 'cat')
````

To quantify the performance of our cat-or-dog classifier, we'll need to split the data into a training and a test set.  We'll also add an index to the dataset, so that we can relate the predictions back to the image urls later on.

````python
from sklearn.cross_validation import train_test_split

index_column = np.vstack(np.arange(data['features'].shape[0]))
indexed_data = np.concatenate([index_column, data['features']], axis=1).astype('int32')

training_features, test_features, training_labels, test_labels = train_test_split(
    indexed_data,
    is_cat,
    test_size=0.1
)

# now remove the index column before we pass the data to the classifier
training_index = training_features[:,0]
training_features = training_features[:,1:]
test_index = test_features[:,0]
test_features = test_features[:,1:]
````

### 3.2 Train a Naive Bayes classifier on the images

Let's start by using a very simple and fast classifer - Naive Bayes

````python
from sklearn.naive_bayes import MultinomialNB

def classify(model, **kwargs):
    classifier = model(**kwargs)
    print "\nTraining:"
    %time classifier.fit(training_features, training_labels)
    
    print "Predicting:"
    cat_predictions = classifier.predict(test_features)
    
    return classifier, cat_predictions
 
classifier, cat_predictions = classify(MultinomialNB)
````

The model has been trained on 25,000 images in just over a second.  Now let's use our trained classifer to make some predictions about the data which it has not seen

````python
%time is_cat_predictions = classifier.predict(test_features)
````

### 3.2 Evaluate Model performance

#### Classification errors

To understand how well our cat-or-dog classifier is performing, let's look at the confusion matrix

````python
%matplotlib inline
from matplotlib import pyplot as plt
from sklearn.metrics import confusion_matrix
 
def plot_confusion_matrix(plot=True):
    confusion = confusion_matrix(test_labels, is_cat_predictions)
    
    if plot:
        fig = plt.figure()
        ax = fig.add_subplot(111)
        ax.matshow(confusion)
        ax.set_xticklabels(['', 'cat', 'dog'])
        ax.set_yticklabels(['', 'cat', 'dog'])
        plt.title('Confusion matrix')
        plt.ylabel('True label')
        plt.xlabel('Predicted label')
        plt.show()
    
    print "\nConfusion matrix:"
    print confusion
    print "\ncats are correctly predicted {0:.2f}% of the time".format((100.0*confusion[0,0])/confusion[0].sum())
    print "dogs are correctly predicted {0:.2f}% of the time".format((100.0*confusion[1,1])/confusion[1].sum())
 
plot_confusion_matrix()
````

If we finetune - ie. retrain the neural network - we can potentially get better performance, but
at the computational cost of having to simulate the training phase for a 8 layer convolutional neural network.

By using a simple linear classifier we can train the model on 22.5k images in < 2s on a laptop.
And with linear classifier there are implementations that will allows us to scale out this approach indefinitely
with a cluster of machines.

#### Classification accuracy

Alongside how often the classifier predicts the correct class, we can measure how accurate it is in it's estimation of class membership

````python
from sklearn.metrics import r2_score
 
def compute_r2_score():
    cat_probability = classifier.predict_proba(test_features)
    r2 = r2_score(test_labels.astype('float32'), cat_probability[:,1])
    print "\nr2 score: {0:.4f}".format(r2)
 
compute_r2_score()
````

### 3.4 Using more powerful models

That's a good start, but can we get better results by using more powerful linear models?

#### Stochastic gradient descent classifier

````python
from sklearn.linear_model import SGDClassifier

classifier, cat_predictions = classify(SGDClassifier, loss='log')
plot_confusion_matrix(plot=False)
compute_r2_score()
````

#### Linear Support Vector Machine

````python
from sklearn.svm import SVC
 
classifier, cat_predictions = classify(SVC, C=1.0, gamma=0.0, kernel='linear', probability=True)
plot_confusion_matrix(plot=False)
compute_r2_score()
````

## 4. Conclusions

## Appendix A:  Understanding model performance

Before diving in and trying to make improvements to the model, let's look at some of the images for which the classifier is making good and bad predictions

````python
cat_probability = classifier.predict_proba(test_features)[:,1]

prediction_error = np.power(test_labels.astype('float32') - cat_probability, 2)

bad_predictions = np.argpartition(prediction_error, -25)[-25:]
good_predictions = np.argpartition(prediction_error, 25)[-25:]

bad_prediction_urls = data['processed_urls'][test_index[bad_predictions]]
good_prediction_urls = data['processed_urls'][test_index[good_predictions]]
````

### Impact of image resolution

````python
import skimage.io

def image_size(url):
    return skimage.io.imread(url).shape[:2]
 
bad_image_sizes = [image_size(url) for url in bad_prediction_urls]
good_image_sizes = [image_size(url) for url in good_prediction_urls]
 
print "Bad predictions [Width, Height]: {0}".format(np.mean(bad_image_sizes, axis=0))
print "Good predictions [Width, Height]: {0}".format(np.mean(good_image_sizes, axis=0))
````

so the network tends to perform more poorly when predicting from lower resolution images.  We can try and improve our model, but this won't help us much if the signal in the input data isn't great to begin with.

### Impact of image cropping

let's center crop the images, so we can see how they look when presented to the network

````python
from scipy.ndimage import zoom
 
def center_crop_image(url):
    image = skimage.io.imread(url)
    
    # resize the image
    input_image_dimensions = np.array([227, 227])
    scale = input_image_dimensions[0].astype('float32') / np.array(image.shape)[:2]
    if scale.min() <= input_image_dimensions.min():
        zoom_factor = scale.max()
    else:
        zoom_factor = scale.min()
    resized_image = zoom(image, (zoom_factor, zoom_factor, 1), order=1)
    
    # centre crop the image
    center = np.array(resized_image.shape[:2]) / 2.0
    crop = np.tile(center, (1, 2))[0] + np.concatenate([-input_image_dimensions / 2.0, input_image_dimensions / 2.0])
    center_cropped_image = resized_image[crop[0]:crop[2], crop[1]:crop[3], :]
    return center_cropped_image
 
bad_images = [center_crop_image(url) for url in bad_prediction_urls]
good_images = [center_crop_image(url) for url in good_prediction_urls]
````

````python
for prediction_type, images in zip(["Bad", "Good"], [bad_images, good_images]):
    fig, axes = plt.subplots(nrows=5, ncols=5, figsize=(12, 12))
    fig.suptitle("{0} predictions".format(prediction_type), fontsize=20)
    for i, image in enumerate(images):
        row = i/5
        column = i % 5
        axes[row, column].xaxis.set_visible(False)
        axes[row, column].yaxis.set_visible(False)
        axes[row, column].imshow(image)
````

a quick non-scientific eye-ball of the data seems to indicate the network performs poorly on noisy, low-light, or out-of-focus images as would a human.
 
There are also a number of images where the face of the cat or dog does not appear in the centre of the image.  As the images are centre-cropped before being presented to the network, this will mean that the network does not get to see some key features that would aid correct classification.  This can be mitigated by oversampling the image (generating multiple crops of the images either side of the centre).
 
Also dogs with pointy-ears tend to get misclassified as cats!
